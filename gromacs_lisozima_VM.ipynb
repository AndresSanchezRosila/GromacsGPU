#**Cargar archivo inicial en carpeta content**

-f indica que el input es un archivo .pdb 
-o indica el output sera archivo  .gro


```
!gmx pdb2gmx -f 1AKI_clean.pdb -o 1AKI_processed.gro -water spce
#usar 15 de force field OPLS
```

editconf converts generic structure format to .gro

 -bt box indica tipo de caja cubica


```
!gmx editconf -f 1AKI_processed.gro -o 1AKI_newbox.gro -c -d 1.0 -bt cubic
```

Solvate puede:
1) Genere una caja de solvente.
2) Solvate una configuración de soluto, p. una proteína, en un baño de moléculas de disolvente.

-cp input con la proteina
-cs input con la config de spc

-o archivo output arch gromacs
-output archivo de topologia


```
!gmx solvate -cp 1AKI_newbox.gro -cs spc216.gro -o 1AKI_solv.gro -p topol.top
```

Cargar ions.mdp a carpeta content

gmx grompp (el preprocesador de gromacs) lee un archivo de topología molecular, comprueba la validez del archivo, expande la topología de una descripción molecular a una descripción atómica.

-f input
-c archivo de estructura
-p archivo de topologia
-o archivo portable de corrida

```
!gmx grompp -f ions.mdp -c 1AKI_solv.gro -p topol.top -o ions.tpr
```

gmx genion reemplaza aleatoriamente moléculas de solvente con iones monoatómicos.

-s archivo input de corrida
-o archivo de esctructura
-p archivo de topologia
-pname nombre del ion positivo NA sodio
-nname nombre del ion negativo CL cloro
-neutral anade suficientes iones para neutralizar el sistema



```
!gmx genion -s ions.tpr -o 1AKI_solv_ions.gro -p topol.top -pname NA -nname CL -neutral
#13 0

```

Cargar minim.mdp a carpeta content

```
!gmx grompp -f minim.mdp -c 1AKI_solv_ions.gro -p topol.top -o em.tpr
```

gmx mdrun es el principal motor de química computacional dentro de GROMACS. Obviamente, realiza simulaciones de dinámica molecular, pero también puede realizar dinámica estocástica, minimización de energía, inserción de partículas de prueba o (re) cálculo de energías.

-v loud and noisi
-deffnm define el nombre de default 

```
!gmx mdrun -v -deffnm em
```

gmx energy extrae componentes energéticos de un archivo de energía.


```
!gmx energy -f em.edr -o potential.xvg
#10 0

```
